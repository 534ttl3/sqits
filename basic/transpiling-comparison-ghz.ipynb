```python
import numpy as np
from qiskit import *
from qiskit.visualization import plot_state_city
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor
from qiskit import Aer  # local simulator
from qiskit import IBMQ  # remote real device
%matplotlib inline
```

```python
ghz = QuantumCircuit(5, 5)
ghz.h(0)
for idx in range(1,5):
    ghz.cx(0,idx)
ghz.barrier(range(5))
ghz.measure(range(5), range(5))
ghz.draw(output='mpl')
```

```python
# ------- Aer qasm simulator  ---------
qasm_counts = execute(ghz, Aer.get_backend('qasm_simulator'), shots=1024).result().get_counts(ghz)

print("qasm simulator: measurement histogram:", qasm_counts); plot_histogram(qasm_counts)
```

```python
# ----- physical device: ibmqx2 ---------
IBMQ.load_account()
```

```python
provider = IBMQ.get_provider(group='open')
provider.backends()
```

```python
from qiskit.visualization import plot_gate_map
backend = provider.get_backend('ibmqx2')
plot_gate_map(backend, plot_directed=True, line_width=2)
```

```python
ghz_circ_simulator = compiler.transpile(ghz, Aer.get_backend('qasm_simulator'))
ghz_circ_simulator.draw(output='mpl')
```

```python
auto_mapped_ghz_circ = compiler.transpile(ghz, provider.get_backend('ibmqx2'), optimization_level=3)
auto_mapped_ghz_circ.draw(output='mpl')
```

```python
custom_mapped_ghz_circ = compiler.transpile(ghz, provider.get_backend('ibmqx2'),
                                            initial_layout=[4,1,2,3,0], optimization_level=0)
custom_mapped_ghz_circ.draw(output='mpl')
```

```python
auto_job_exp = execute(auto_mapped_ghz_circ, backend=provider.get_backend('ibmqx2'))
job_monitor(auto_job_exp)
print("Exp. job ID:", auto_job_exp.job_id())
```

```python
custom_job_exp = execute(custom_mapped_ghz_circ, backend=provider.get_backend('ibmqx2'))
job_monitor(custom_job_exp)
print("Exp. job ID:", custom_job_exp.job_id())
```

```python
plot_histogram([auto_job_exp.result().get_counts(), 
                custom_job_exp.result().get_counts(), 
                qasm_counts],
               legend=['ibmqx2, optimized', 'ibmqx2, manually mapped', 'Aer qasm simulator backend'], figsize=[4*2.5, 3*2.5])
```
