```python
import numpy as np
from qiskit import *
from qiskit.visualization import plot_state_city
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor
from qiskit import Aer  # local simulator
from qiskit import IBMQ  # remote real device
%matplotlib inline

# --------------- XX stabilizer measurement circuit -----------
# detect a phase flip (Z_1 or Z_2) on one of the physical qubits of the logical qubit

circ = QuantumCircuit(3,1)
circ.h(0)
circ.h(1)
circ.h(2)
circ.z(0)
circ.cx(2, 0)
circ.cx(2, 1)
circ.h(2)

circ.barrier()
circ_copy = circ.copy()

# measure ancilla qubit q_2 into bit c_0
circ.measure(2,0)

circ.draw()
```

```python
air_statev_job = execute(circ_copy, Aer.get_backend('statevector_simulator'))
air_statev_result = air_statev_job.result()

# get complex amplitudes as array of size 2^n where 
# 0 -> <0_n ... 0_0|\psi>, 1 -> <0_n ... 1_0|\psi>, ...
air_output_state = air_statev_result.get_statevector(circ, decimals=3) 
print("statevector before measurement:", air_output_state)
```

```python
plot_state_city(air_output_state)
```

```python
# ------- Aer qasm simulator  ---------
qasm_counts = execute(circ, Aer.get_backend('qasm_simulator'), shots=1024).result().get_counts(circ)

print("qasm simulator: measurement histogram:", qasm_counts); plot_histogram(qasm_counts)
```

```python
# ----- physical device: ibmqx2 ---------
IBMQ.load_account()
IBMQ.providers()  # list all providers
ibmq_provider_backend = IBMQ.get_provider(group='open').get_backend('ibmqx2')  # get device as backend
```

```python
# run the circuit and monitor the job
job_exp = execute(circ, backend=ibmq_provider_backend)
job_monitor(job_exp)  # async
print("Exp. job ID:", job_exp.job_id())
```

```python
result_exp = job_exp.result()
counts_exp = result_exp.get_counts(circ)
plot_histogram([counts_exp,qasm_counts], legend=['ibmqx2 backend', 'Aer qasm simulator backend'])
```
